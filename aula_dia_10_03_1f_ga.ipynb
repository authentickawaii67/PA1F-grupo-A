calcular a área do retângulo

```python
# entrada
comprimento = float(input("digite o comprimento do retângulo:"))
altura = float(input("digite a altura do retângulo"))

# processamento

area = comprimento * altura


# saída
print(f"o total da área to retângulo é:{area}")
```

fazer um programa que calcule o dobro de um número

```python
# entrada

numero = float(input("digite o número que deseja saber o seu dobro:"))

# processamento

resultado = numero * 2

# saida

print (f"o dobro de {numero} é igual a {resultado}.")
```

escrever um programa que calcula a média das notas do aluno

entrada: nome, idade e as 4 notas do bimestre

processamento: calcular a media

saida: mostrar o nome do aluno e a media obtida

```python
# entrada

nome = str(input("digite seu nome: "))
idade = float(input("digite sua idade: "))
nota1 = float(input("adicione sua primeira nota do bimestre:"))
nota2 = float(input("adicione sua segunda nota do bimestre:"))
nota3 = float(input("adicione sua terceira nota do bimestre:"))
nota4 = float(input("adicione sua quarta nota do bimestre:"))

# processamento

media = (nota1 + nota2 + nota3 + nota4) /4

# saida

print (f"olá {nome}, sua média é igual a {media}.")
```
