```python
# questao 2

# Crie um programa que leia o valor de uma compra e aplique o desconto conforme as regras:
# se o valor é até R$100: sem desconto
# se o valor for de R$101 a R$500: 10% de desconto
# se o valor for acima de R$500: 20% de desconto
# no final, mostre o valor final da compra.

# entrada

valor = float(input ("digite o valor total da sua compra:"))

# processamento / saida

if valor <= 100:
  valor_final = valor
  print(f"o valor final da compra é R${valor_final:.2f} (sem desconto).")

elif valor <= 500:
  desconto = valor * 0.10
  valor_final = valor - desconto
  print(f"o valor final da compra é R${valor_final:.2f} (10% de desconto).")

else:
  desconto = valor * 0.20
  valor_final = valor - desconto
  print(f"o valor final da compra é R${valor_final:.2f} (20% de desconto).")
```

```python
# questao 4

# entrada

energia = int(input("digite a energia do seu personagem: "))

# processamento / saida

if energia <= 0:
  print ("seu personagem está morto.")

elif energia >= 1 and energia <= 30:
  print ("seu personagem está com a energia crítica.")

elif energia >= 31 and energia <= 70:
  print ("seu personagem está com a energia normal.")

elif energia >= 71 and energia <= 100:
  print ("seu personagem está saudavel.")
else:
  print("Valor de energia inválido.")
```

```python
# questao 5

# Crie um programa que leia a média final de um aluno e classifique:
# se a média é maior ou igual a 7: Aprovado
# se a média é entre 5 e 6.9: Recuperação
# se a média é menor que 5: reprovado

# entrada

nome = str(input("digite seu nome: "))
nota1 = float(input("adicione sua primeira nota do bimestre:"))
nota2 = float(input("adicione sua segunda nota do bimestre:"))
nota3 = float(input("adicione sua terceira nota do bimestre:"))
nota4 = float(input("adicione sua quarta nota do bimestre:"))

# processamento / saida

media = (nota1 + nota2 + nota3 + nota4) /4

if media >= 7:
  print(f"{nome}, sua média é {media:.2f}, você está aprovado!")

elif media >= 5 and media < 7:
  print(f"{nome}, sua média é {media:.2f}, você está de recuperação.")

else:
  print(f"{nome}, sua média é {media:.2f}, você está reprovado.")
```
