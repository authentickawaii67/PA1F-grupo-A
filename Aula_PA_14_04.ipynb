```python
# Peça a nota de um aluno e informe:
#	Aprovado (nota ≥ 7)
#	Recuperação (5 ≤ nota < 7)
#	Reprovado (nota < 5)

# entrada

nota = int(input("Digite sua nota: "))
nome = str(input("Digite seu nome: "))

# processamento/saida

if nota >= 7:
  print(f"{nome}, sua média é {nota:.2f}, você está aprovado!")

elif nota >= 5 and nota < 7:
  print(f"{nome}, sua média é {nota:.2f}, você está de recuperação.")

elif nota < 5:
  print(f"{nome}, sua média é {nota:.2f}, você está reprovado.")
```

```python
# Peça dois números e uma operação (+, -, *, /). Use if/elif para calcular o resultado.

# entrada

num1 = float(input("Digite o primeiro número: "))
operacao = input("Digite a operação (+, -, *, /): ")
num2 = float(input("Digite o segundo número: "))

# aaaaa

if operacao == '+':
    resultado = num1 + num2
    print(f"Resultado: {num1} + {num2} = {resultado}")

elif operacao == '-':
    resultado = num1 - num2
    print(f"Resultado: {num1} - {num2} = {resultado}")

elif operacao == '*':
    resultado = num1 * num2
    print(f"Resultado: {num1} * {num2} = {resultado}")

elif operacao == '/':
    if num2 != 0:
        resultado = num1 / num2
        print(f"Resultado: {num1} / {num2} = {resultado}")
    else:
        print("Erro: Divisão por zero não permitida.")

else:
    print("Operação inválida. Use +, -, *, ou /.")
```

```python
# Peça um número de 1 a 7 e mostre o dia correspondente (ex: 1 = domingo).

# e

num = int(input("Digite um número de um dia da semana: "))

# p s

if num == 1:
  print(f"{num} = domingo")

elif num == 2:
  print(f"{num} = segunda")

elif num == 3:
  print(f"{num} = terça")

elif num == 4:
  print(f"{num} = quarta")

elif num == 5:
  print(f"{num} = quinta")

elif num == 6:
  print(f"{num} = sexta")

elif num == 7:
  print(f"{num} = sábado")

else:
  print(f"Esse não é um némero de um dia da semana válido.")
```
