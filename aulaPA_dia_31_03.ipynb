```python
# eestrutura de decisoões

# escrever um programa que recebe o nome e a idade do usuario e verifica se o usuario é criança, adolescente ou adulto

# >= 12 anos é criança
# =< 18 anos é adolescente
# adulto

# entrada

nome = str(input ("digite o seu nome: "))
idade = int(input("digite sua idade: "))

# processamento / saida

if idade <= 12:
  print (f"{nome}, voce tem {idade} anos, voce é uma criança.")

elif idade <= 18:
  print (f"{nome}, voce tem {idade} anos, voce é um adolescente.")

else:
  print (f"{nome}, voce tem {idade} anos, voce é um adulto.")


```

```python
# em um jogo, um personagem possui uma variavel chamada energia. O progarama deve:
# se a energia for maior que 70 - forte, se a energia estiver entre 30 e 70 - cansado
# se a energia estiver menor que 30 - fraco.

# entrada

energia = int(input("digite a energia do seu personagem: "))

# processamento / saida

if energia <= 30:
  print ("seu personagem está fraco.")

if energia <= 70:
  print ("seu personagem está cansado.")

if energia > 70:
  print ("seu personagem está forte.")
```
