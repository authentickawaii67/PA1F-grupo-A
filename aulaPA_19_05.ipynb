```python
# escrever um programa que recebe as notas dos 4 bimestres e imprime a media com
# o resultado de aprovado, recuperaçao ou reprovado

# regras

# ultilizar while
# nota igual ou maior que 7 aprovado
# nota igual ou maior que 5 recuperaçao
# nota menor que 5 reprovado

notas = 0
bimestres = 0

while bimestres < 4:

    try:
        nota = float(input(f"Digite a nota do {bimestres + 1}º bimestre: "))
        if 0 <= nota <= 10:
            notas += nota
            bimestres += 1

        else:
            print("Por favor, digite uma nota entre 0 e 10.")

    except ValueError:
        print("Entrada inválida. Por favor, digite um número para a nota.")

media = notas / 4

print(f"\nMédia final: {media:.2f}")

if media >= 7:
    print("Você foi aprovado!")

elif media >= 5:
    print("Você está de recuperação.")

else:
    print("Você foi reprovado.")
```

```python
# Você foi contratado para desenvolver um sistema simples de caixa eletrônico.
# O sistema deverá funcionar através de um menu interativo utilizando “while”
# O programa de iniciar com um saldo inicial de R$1000,00 e permitir que o
# usuário escolha diferentes operações até decidir sair do sistema

# Opções:

#	Sacar dinheiro
#	Depositar dinheiro
#	Ver saldo
#	Sair do banco


saldo = 1000.00

while True:
    print("\n--- Caixa Eletrônico ---")
    print("1. Sacar dinheiro")
    print("2. Depositar dinheiro")
    print("3. Ver saldo")
    print("4. Sair do banco")

    try:
        opcao = int(input("Escolha uma opção: "))

        if opcao == 1: # Sacar dinheiro
            valor_saque = float(input("Digite o valor a sacar: R$"))
            if valor_saque > saldo:
                print("Saldo insuficiente!")
            elif valor_saque <= 0:
                print("Valor de saque inválido.")
            else:
                saldo -= valor_saque
                print(f"Saque de R${valor_saque:.2f} realizado com sucesso. Novo saldo: R${saldo:.2f}")

        elif opcao == 2: # Depositar dinheiro
            valor_deposito = float(input("Digite o valor a depositar: R$"))
            if valor_deposito <= 0:
                print("Valor de depósito inválido.")
            else:
                saldo += valor_deposito
                print(f"Depósito de R${valor_deposito:.2f} realizado com sucesso. Novo saldo: R${saldo:.2f}")

        elif opcao == 3: # Ver saldo
            print(f"Seu saldo atual é: R${saldo:.2f}")

        elif opcao == 4: # Sair do banco
            print("Obrigado por usar nosso caixa eletrônico. Até logo!")
            break

        else:
            print("Opção inválida. Por favor, escolha uma opção entre 1 e 4.")

    except ValueError:
        print("Entrada inválida. Por favor, digite um número correspondente à opção.")
```
