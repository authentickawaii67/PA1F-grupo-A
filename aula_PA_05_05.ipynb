```python
# Escrever um programa que o usuário digita um número de 1 a 10 e o programa impri a tabuada do número digitado.

numero_tabuada = int(input("Digite um número de 1 a 10 para ver a tabuada: "))

if 1 <= numero_tabuada <= 10:
    print(f"Tabuada do {numero_tabuada}:")
    for i in range(1, 11):
        resultado = numero_tabuada * i
        print(f"{numero_tabuada} x {i} = {resultado}")
else:
    print("Por favor, digite um número entre 1 e 10.")
```

```python
# 1 - Escrever um programa que o usuário digita um número e o programa imprimi até o número digitado.

numero = int(input("Digite um número: "))

print(f"Números de 0 até {numero}:")
for i in range(numero + 1):
    print(i)
```

```python
# 2 - Escrever um programa que o usuário digita um número e o programa imprimi o número digitado até o zero (contagem regressiva).

numero = int(input("Digite um número para iniciar a contagem regressiva: "))

print(f"Contagem regressiva a partir de {numero}:")
for i in range(numero, -1, -1):
    print(i)
```

```python
# 3 - Escrever um programa que o usuário digita um número e o programa imprimi os números pares.

numero = int(input("Digite um número: "))

print(f"Números pares de 0 até {numero}:")
for i in range(numero + 1):
    if i % 2 == 0:
        print(i)
```

```python
# 4 - Escrever um programa que o usuario digita um numero e programa soma os números até o numero digitado.

numero = int(input("Digite um número inteiro positivo: "))

soma = 0
for i in range(1, numero + 1):
    soma += i

print(f"A soma dos números de 1 até {numero} é: {soma}")
```

```python
# 5 - Escrever um programa que o usuário digita um número e programa soma os números até o número digitado.

numero = int(input("Digite um número: "))
soma = 0

print(f"Números somados até {numero}:")
for i in range(numero + 1):
  soma = soma + i
print(soma)
```

```python
# 6 - Escrever um programa que o usuário digita uma palavra e o programa conta as letras da palavra.

palavra = input("Digite uma palavra: ")

quantidade_letras = len(palavra)
print(f"A palavra '{palavra}' tem {quantidade_letras} letras.")
```

```python
# 7 - Escrever um programa que o usuário digita uma palavra e o programa conta as vogais da palavra.

palavra = input("Digite uma palavra: ")

palavra_minuscula = palavra.lower()
vogais = "aeiou"
quantidade_vogais = 0

for letra in palavra_minuscula:
    if letra in vogais:
        quantidade_vogais += 1

print(f"A palavra '{palavra}' tem {quantidade_vogais} vogais.")
```

```python
# 8 - Escrever um programa que o usuário digita uma palavra e o programa imprimi a palavra invertida.

palavra = input("Digite uma palavra: ")
invertida = ""

for i in (palavra):

  invertida = i + invertida
print(f"A palavra {palavra} invertida é {invertida}.")
```

```python
# 9 - Escrever um programa que o usuário digita 04 números e o programa imprimi o maior.

print("Digite quatro números para encontrar o maior:")
numero1 = float(input("Número 1: "))
numero2 = float(input("Número 2: "))
numero3 = float(input("Número 3: "))
numero4 = float(input("Número 4: "))

maior_numero = max(numero1, numero2, numero3, numero4)

print(f"O maior número digitado é: {maior_numero}")
```
