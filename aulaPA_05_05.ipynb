```python
# Escrever um programa que o usuário digita um número de 1 a 10 e o programa impri a tabuada do número digitado.

numero_tabuada = int(input("Digite um número de 1 a 10 para ver a tabuada: "))

if 1 <= numero_tabuada <= 10:
    print(f"Tabuada do {numero_tabuada}:")
    for i in range(1, 11):
        resultado = numero_tabuada * i
        print(f"{numero_tabuada} x {i} = {resultado}")
else:
    print("Por favor, digite um número entre 1 e 10.")
```

```python
# 7 - Escrever um programa que o usuário digita uma palavra e o programa conta as vogais da palavra.

palavra = input("Digite uma palavra: ")

palavra_minuscula = palavra.lower()
vogais = "aeiou"
quantidade_vogais = 0

for letra in palavra_minuscula:
    if letra in vogais:
        quantidade_vogais += 1

print(f"A palavra '{palavra}' tem {quantidade_vogais} vogais.")
```

*   `" ".join(map(str, last_row_numbers))`: Converte cada número para string e os junta com espaços, criando a representação da linha como texto (ex: `'5 4 3 2 1 2 3 4 5'`).

**3. Loop para Cada Linha da Pirâmide (`for i in range(1, n + 1):`):**

*   Este loop executa `n` vezes, uma vez para cada linha da pirâmide. A variável `i` representa o número da linha atual (de `1` a `n`).

```python
    left_part = list(range(i, 0, -1)) # ex: para i=3 -> [3, 2, 1]
    right_part = list(range(2, i + 1)) # ex: para i=3 -> [2, 3]
    row_numbers = left_part + right_part
```

**4. Construindo os Números da Linha Atual:**

*   Dentro do loop, para cada `i` (linha atual), criamos os números que compõem aquela linha.
*   `left_part`: Gera a sequência decrescente de `i` até `1`.
*   `right_part`: Gera a sequência crescente de `2` até `i`.

```python
    row_string = " ".join(map(str, row_numbers))
```

**5. Formatando a Linha como String:**

*   `map(str, row_numbers)`: Converte cada número na lista `row_numbers` para uma string.

```python
    current_width = len(row_string)
    padding = (max_width - current_width) // 2
```

**6. Calculando o Preenchimento para Centralização:**

*   `current_width`: Obtém o comprimento da `row_string` atual.
*   `padding`: Calcula quantos espaços em branco devem vir antes da `row_string` para que ela fique centralizada. A diferença entre `max_width` e `current_width` é dividida por 2 (com `//` para divisão inteira) para distribuir os espaços igualmente em ambos os lados.

```python
    print(" " * padding + row_string)
```

**7. Imprimindo a Linha Centralizada:**

*   `" " * padding`: Cria uma string com a quantidade necessária de espaços em branco.
*   `... + row_string`: Concatena esses espaços com a `row_string`.
*   `print(...)`: Imprime a linha final, agora centralizada.

Este método garante que a pirâmide seja sempre centralizada, independentemente do número `N` digitado, porque ele calcula dinamicamente a largura da última linha e usa essa largura como referência para todas as outras.
