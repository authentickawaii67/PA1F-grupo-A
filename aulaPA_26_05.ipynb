```python
# calculadora com as 4 opçoes

# menu

# opçao 1 - somar
# opçao 2 - subtrair
# opçao 3 - multiplicar
# opçao 4 - divisao
# opçao 5 - sair

while True:
    print("\n--- Calculadora ---")
    print("1. Somar")
    print("2. Subtrair")
    print("3. Multiplicar")
    print("4. Dividir")
    print("5. Sair")

    try:
        opcao = int(input("Escolha uma opção: "))

        if opcao == 5:
            print("Saindo da calculadora. Até logo!")
            break

        if opcao < 1 or opcao > 4:
            print("Opção inválida. Por favor, escolha uma opção entre 1 e 5.")
            continue

        num1 = float(input("Digite o primeiro número: "))
        num2 = float(input("Digite o segundo número: "))

        if opcao == 1:
            resultado = num1 + num2
            print(f"Resultado: {num1} + {num2} = {resultado}")

        elif opcao == 2:
            resultado = num1 - num2
            print(f"Resultado: {num1} - {num2} = {resultado}")

        elif opcao == 3:
            resultado = num1 * num2
            print(f"Resultado: {num1} * {num2} = {resultado}")

        elif opcao == 4:
            if num2 == 0:
                print("Erro: Divisão por zero não é permitida.")
            else:
                resultado = num1 / num2
                print(f"Resultado: {num1} / {num2} = {resultado}")

    except ValueError:
        print("Entrada inválida. Por favor, digite números para as operações e um número para a opção.")
```

```python
# faça um jogo inteligente onde o usuario digita 2 numeros e decide oque fazer
# escolhendo uma opçao no menu

# menu

# 1 - par/impar
# 2 - positivo
# 3 - maior dos 2 numero
# 4 - sair

while True:
    print("\n--- Jogo Inteligente ---")
    print("1 - Par/Ímpar")
    print("2 - Positivo")
    print("3 - Maior dos 2 números")
    print("4 - Sair")

    try:
        opcao = int(input("Escolha uma opção: "))

        if opcao == 4:
            print("Saindo do jogo. Até mais!")
            break

        if opcao < 1 or opcao > 3:
            print("Opção inválida. Por favor, escolha uma opção entre 1 e 4.")
            continue

        num1 = float(input("Digite o primeiro número: "))
        num2 = float(input("Digite o segundo número: "))

        if opcao == 1:
            print(f"\nAnálise de Par/Ímpar:")
            if num1 % 2 == 0:
                print(f"O número {num1} é PAR.")
            else:
                print(f"O número {num1} é ÍMPAR.")

            if num2 % 2 == 0:
                print(f"O número {num2} é PAR.")
            else:
                print(f"O número {num2} é ÍMPAR.")

        elif opcao == 2:
            print(f"\nAnálise de Positivo/Negativo:")
            if num1 > 0:
                print(f"O número {num1} é POSITIVO.")
            elif num1 < 0:
                print(f"O número {num1} é NEGATIVO.")
            else:
                print(f"O número {num1} é ZERO.")

            if num2 > 0:
                print(f"O número {num2} é POSITIVO.")
            elif num2 < 0:
                print(f"O número {num2} é NEGATIVO.")
            else:
                print(f"O número {num2} é ZERO.")

        elif opcao == 3:
            print(f"\nComparação de números:")
            if num1 > num2:
                print(f"O maior número é {num1}.")
            elif num2 > num1:
                print(f"O maior número é {num2}.")
            else:
                print("Os dois números são iguais.")

    except ValueError:
        print("Entrada inválida. Por favor, digite números válidos para as operações e um número para a opção.")
```

```python
# escrever um programa que verifica a senha, o usuario tem 3 tentativas, caso erre
# as 3 tentativas imprimir uma mensagem de "Acesso negado. Tente novamente mais tarde."

senha_correta = "6767"
tentativas = 0
tentativas_maximas = 3

while tentativas < tentativas_maximas:
    senha_digitada = input("Digite a senha: ")

    if senha_digitada == senha_correta:
        print("Acesso concedido! Bem-vindo.")
        break
    else:
        tentativas += 1
        tentativas_restantes = tentativas_maximas - tentativas
        if tentativas_restantes > 0:
            print(f"Senha incorreta. Você tem {tentativas_restantes} tentativa(s) restante(s).")
        else:
            print("Acesso negado. Você excedeu o número de tentativas.")
            print("Tente novamente mais tarde.")
```
