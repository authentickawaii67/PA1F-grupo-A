```python
saldo = 1000.00

while True:
    print("\n--- Caixa Eletrônico ---")
    print("1. Sacar dinheiro")
    print("2. Depositar dinheiro")
    print("3. Ver saldo")
    print("4. Sair do banco")

    try:
        opcao = int(input("Escolha uma opção: "))

        if opcao == 1:
            try:
                valor_saque = float(input("Digite o valor a sacar: R$"))
                if valor_saque > saldo:
                    print("Saldo insuficiente!")
                elif valor_saque <= 0:
                    print("Valor de saque inválido.")
                else:
                    saldo -= valor_saque
                    print(f"Saque de R${valor_saque:.2f} realizado com sucesso. Novo saldo: R${saldo:.2f}")
            except ValueError:
                print("Entrada inválida. Por favor, digite um valor numérico para o saque.")

        elif opcao == 2:
            try:
                valor_deposito = float(input("Digite o valor a depositar: R$"))
                if valor_deposito <= 0:
                    print("Valor de depósito inválido.")
                else:
                    saldo += valor_deposito
                    print(f"Depósito de R${valor_deposito:.2f} realizado com sucesso. Novo saldo: R${saldo:.2f}")
            except ValueError:
                print("Entrada inválida. Por favor, digite um valor numérico para o depósito.")

        elif opcao == 3:
            print(f"Seu saldo atual é: R${saldo:.2f}")

        elif opcao == 4:
            print("Obrigado por usar nosso caixa eletrônico. Até logo!")
            break

        else:
            print("Opção inválida. Por favor, escolha uma opção entre 1 e 4.")

    except ValueError:
        print("Entrada inválida. Por favor, digite um número correspondente à opção.")
```
