```python
saldo = 1000.00

while True:
    print("\n--- Caixa Eletrônico ---")
    print("1. Sacar dinheiro")
    print("2. Depositar dinheiro")
    print("3. Ver saldo")
    print("4. Sair do banco")

    try:
        opcao = int(input("Escolha uma opção: "))

        if opcao == 1:
            try:
                valor_saque = float(input("Digite o valor a sacar: R$"))
                if valor_saque > saldo:
                    print("Saldo insuficiente!")
                elif valor_saque <= 0:
                    print("Valor de saque inválido.")
                else:
                    saldo -= valor_saque
                    print(f"Saque de R${valor_saque:.2f} realizado com sucesso. Novo saldo: R${saldo:.2f}")
            except ValueError:
                print("Entrada inválida. Por favor, digite um valor numérico para o saque.")

        elif opcao == 2:
            try:
                valor_deposito = float(input("Digite o valor a depositar: R$"))
                if valor_deposito <= 0:
                    print("Valor de depósito inválido.")
                else:
                    saldo += valor_deposito
                    print(f"Depósito de R${valor_deposito:.2f} realizado com sucesso. Novo saldo: R${saldo:.2f}")
            except ValueError:
                print("Entrada inválida. Por favor, digite um valor numérico para o depósito.")

        elif opcao == 3:
            print(f"Seu saldo atual é: R${saldo:.2f}")

        elif opcao == 4:
            print("Obrigado por usar nosso caixa eletrônico. Até logo!")
            break

        else:
            print("Opção inválida. Por favor, escolha uma opção entre 1 e 4.")

    except ValueError:
        print("Entrada inválida. Por favor, digite um número correspondente à opção.")
```

```python
# fazer  um jogo de adivinhação

#o usuario ira digitar um numero de 0 a 10 para adivinhar o numero secreto
#o usuario tera 3 tentativas para acertar o numero
#o programa dever imprimir uma mensagem se o palpite do usuario e maior ou menor que o numero secreto
#o programa devera imprimir uma mensagem de quente, morno ou frio

# regras

# se a diferença do palpite para o numero secretor for 3 frio, 2-morno e 1-quente

import random
a = random.randint(0, 10)
tentativas = 0
tm = 3

print(f"Tente adivinhar o numero secreto.")

while tentativas < tm:
  try:
    b = int(input(f"Tentativa {tentativas + 1}/{tm}: Digite seu palpite: "))
  except ValueError:
    print(f"Entrada inválida. Por favor, digite um número inteiro.")
    continue

  if not (0 <= b <= 10):
      print(f"Por favor, digite um número entre 0 e 10.")
      continue

  if b == a:
    print(f"Parabéns! Você acertou o número!")
    break

  else:
    tentativas += 1
    tr = tm - tentativas

    if tr > 0:
      if b < a:
        print(f"Seu palpite é menor que o número secreto.")
      else:
        print(f"Seu palpite é maior que o número secreto.")

      diferenca = (a - b)
      if diferenca == 1:
        print(f"Quente!")
      elif diferenca == 2:
        print(f"Morno.")
      else:
        print("Frio.")
      print(f"Você ainda tem {tr} tentativas.")

    else:
      print(f"Suas tentativas acabaram.")
      print(f"O número secreto era {a}.")
      break
```
