```python
# escreve um programa que verifica se a palavra é palindromo e imprime a palavra

palavra = str(input("Digite a palavra: "))
invertida = palavra[::-1]

if palavra == invertida:
    print(f'A palavra "{palavra}" é um palíndromo.')
else:
    print(f'A palavra "{palavra}" não é um palíndromo.')
```

```python
# escrever um programa que o usuario digita um nuemero e o programa imprime os
# numeros pares de 0 ate o numero.

numero = int(input("Digite um número: "))

print(f"Números pares de 0 até {numero}:")
for i in range(numero + 1):
    if i % 2 == 0:
        print(i)
```
